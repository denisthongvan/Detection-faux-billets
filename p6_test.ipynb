```python
import pandas as pd
```

```python
test = pd.read_csv('example.csv')
lr = pd.read_pickle('model.pkl')
```

```python
test
```

```python
X = test.iloc[:,:-1]
```

```python
probability = lr.predict_proba(X.values)[:,1]
```

```python
probability
```

```python
proba = pd.Series(probability.round(3), name='value')
```

```python
test_final=pd.concat([test,proba], axis=1)
```

```python
test_final
```

```python
resultat=[]
for i in test_final.value>0.5:
    if i is True:
        resultat.append('Vrai')
    else:
        resultat.append('Faux')
        
test_final['resultat']=resultat
test_final
```
